```python
# Use 50k stories for training — enough to see clear improvement, trains in ~1.5hrs on Colab T4
train_subset = ds['train'].select(range(50000))
val_subset   = ds['validation'].select(range(2000))

def tokenize(batch):
    tokens = tokenizer(
        batch['text'],
        truncation=True,
        max_length=256,
        padding='max_length',
    )
    tokens['labels'] = tokens['input_ids'].copy()
    return tokens

train_tokenized = train_subset.map(tokenize, batched=True, remove_columns=['text'])
val_tokenized   = val_subset.map(tokenize,   batched=True, remove_columns=['text'])

train_tokenized.set_format('torch')
val_tokenized.set_format('torch')

print(f"Train samples: {len(train_tokenized)}")
print(f"Val samples:   {len(val_tokenized)}")
```

```python
# ============================================================
#  KIDS STORY GPT-2 FINE-TUNING — COMPLETE IN ONE CELL
# ============================================================

# 1. Install
import subprocess
subprocess.run(["pip", "install", "transformers", "datasets", "accelerate", "-q"])

# 2. Mount Drive first so everything saves there
from google.colab import drive
drive.mount('/content/drive')

SAVE_PATH = "/content/drive/MyDrive/tinystories-gpt2-final"

# 3. Imports
from datasets import load_dataset
from transformers import (
    GPT2Tokenizer, GPT2LMHeadModel,
    TrainingArguments, Trainer,
    DataCollatorForLanguageModeling
)
import os, math

# 4. Load dataset
print("Loading dataset...")
ds = load_dataset("roneneldan/TinyStories")
print(f"Train: {len(ds['train'])} | Val: {len(ds['validation'])}")

# 5. Load model
print("Loading GPT-2...")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token
model = GPT2LMHeadModel.from_pretrained("gpt2")
print(f"Parameters: {model.num_parameters():,}")

# 6. Tokenize (50k train, 2k val)
print("Tokenizing...")
def tokenize(batch):
    tokens = tokenizer(batch['text'], truncation=True, max_length=256, padding='max_length')
    tokens['labels'] = tokens['input_ids'].copy()
    return tokens

train_tok = ds['train'].select(range(50000)).map(tokenize, batched=True, remove_columns=['text'])
val_tok   = ds['validation'].select(range(2000)).map(tokenize, batched=True, remove_columns=['text'])
train_tok.set_format('torch')
val_tok.set_format('torch')
print("Tokenized!")

# 7. Train — saves checkpoints directly to Drive
print("Training... (this takes ~1.5 hours, you can close the tab)")
args = TrainingArguments(
    output_dir=SAVE_PATH,
    num_train_epochs=2,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    warmup_steps=200,
    weight_decay=0.01,
    logging_steps=500,
    eval_strategy="epoch",
    save_strategy="epoch",
    save_total_limit=1,
    load_best_model_at_end=True,
    fp16=True,
    report_to="none",
)

trainer = Trainer(
    model=model,
    args=args,
    train_dataset=train_tok,
    eval_dataset=val_tok,
    data_collator=DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False),
)

trainer.train()

# 8. Save final model + tokenizer to Drive
print("Saving to Drive...")
trainer.save_model(SAVE_PATH)
tokenizer.save_pretrained(SAVE_PATH)

# 9. Verify
print("\nFiles saved:")
for f in os.listdir(SAVE_PATH):
    size = os.path.getsize(f"{SAVE_PATH}/{f}")
    print(f"  {f}  ({size/1024/1024:.1f} MB)")

# 10. Evaluate
eval_results = trainer.evaluate()
perplexity = math.exp(eval_results['eval_loss'])
print(f"\nEval Loss:  {eval_results['eval_loss']:.4f}")
print(f"Perplexity: {perplexity:.2f}")

# 11. Quick generation test
from transformers import pipeline
gen = pipeline('text-generation', model=SAVE_PATH, tokenizer=tokenizer)
result = gen(
    "Once upon a time, a little boy named Ali found a mysterious map",
    max_length=150, do_sample=True, temperature=0.8
)
print("\nSample output:")
print(result[0]['generated_text'])

print("\nDONE! Model saved to Google Drive.")
```

```python
# ── Load BASE GPT-2 for comparison ───────────────────────────────────────────
import os
os.environ["HF_HUB_DOWNLOAD_TIMEOUT"] = "120"

base_tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
base_tokenizer.pad_token = base_tokenizer.eos_token
base_model = GPT2LMHeadModel.from_pretrained("gpt2")
base_model.eval()
print("Base model loaded!")
```

```python
# ── Perplexity comparison ────────────────────────────────────────────────────
from datasets import load_dataset

ds = load_dataset("roneneldan/TinyStories")
samples = ds['validation'].select(range(200))

def compute_perplexity(mdl, tok, samples):
    total_loss, count = 0, 0
    with torch.no_grad():
        for item in samples:
            enc = tok(item['text'], return_tensors='pt',
                      truncation=True, max_length=256)
            out = mdl(**enc, labels=enc['input_ids'])
            total_loss += out.loss.item()
            count += 1
    return math.exp(total_loss / count)

print("Computing perplexity (takes ~2 min)...")
base_ppl   = compute_perplexity(base_model,   base_tokenizer, samples)
tuned_ppl  = compute_perplexity(model,        tokenizer,      samples)

print(f"\n{'='*45}")
print(f"  Base GPT-2 perplexity:        {base_ppl:.2f}")
print(f"  Fine-tuned GPT-2 perplexity:  {tuned_ppl:.2f}")
print(f"  Improvement:                  {base_ppl/tuned_ppl:.1f}x better")
print(f"{'='*45}")
print("\n--- COPY THIS FOR YOUR REPORT ---")
print(f"Base GPT-2:        {base_ppl:.2f}")
print(f"Fine-tuned GPT-2:  {tuned_ppl:.2f}")
print(f"Improvement:       {base_ppl/tuned_ppl:.1f}x")
```

```python
# ── Side by side quality comparison ─────────────────────────────────────────
base_gen  = pipeline('text-generation', model=base_model,  tokenizer=base_tokenizer)
tuned_gen = pipeline('text-generation', model=model,       tokenizer=tokenizer)

prompts = [
    "Once upon a time, a little boy named Ali found a mysterious map",
    "Sara wanted to keep the baby bird forever, but",
    "The dragon offered Omar great power, but",
]

for prompt in prompts:
    print("=" * 55)
    print(f"PROMPT: {prompt}")
    print("-" * 55)
    r1 = base_gen(prompt,  max_length=120, do_sample=True, temperature=0.8)
    r2 = tuned_gen(prompt, max_length=120, do_sample=True, temperature=0.8)
    print(f"BASE GPT-2:\n{r1[0]['generated_text']}\n")
    print(f"FINE-TUNED:\n{r2[0]['generated_text']}\n")
```
